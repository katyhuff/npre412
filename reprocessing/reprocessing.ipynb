### What are actinides?

Actinides are all elements that are heavier than Actinium (Z=89). We typically care about a particular set of actinides called _transuranics_ or TRU. These are all elements heavier than Uranium (Z=92). Particularly:

- Np (Z=93)
- Pu (Z=94)
- Am (Z=95)
- Cm (Z=96)


### Management of Actinides

There are three primary reasons to manage actinides. 

- Repository benefit: The decay heat burden of the actinides is a challenge for geologic disposal and increases repository management horizon.
- Energy Recovery: There is energy stored in many of these actinides, as some are fissile or have fissile daughters.
- Proliferation Resistence: Actinides can be used to construct a nuclear weapon (specifically, Pu).

## PUREX

PUREX is simple, conventional separation of FP from U-Pu-MA (Does not reduce radioactivity but separates for further processing). [LLNL Report source material here.](https://e-reports-ext.llnl.gov/pdf/799624.pdf)

- Head end: (Sometimes Voloxidation), Chopping, shearing, dissolving.
- 1st Processing Phase: Coextraction. "In the first cycle, uranium and plutonium are co-extracted into an
organic phase, leaving the fission products and other actinide activation products in the aqueous phase"
- Second Processing Phase: Pu Extraction. "In the second cycle, plutonium is stripped away from the uranium and
returned to an aqueous phase,"
- Third Processing Phase: "In the third cycle, uranium is stripped away from the organic phase into dilute nitric acid"
- Finally: Additional purification and return to appropriate valences.

# Fuel Utilization

One reason to consider reprocessing is economic. While right now, the cost of uranium and associated enrichment services is low in comparison to other fuel forms, per kWh, this may not always be the case. 

\begin{align}
U &= \frac{\mbox{mass of fuel consumed}}{\mbox{mass of resource (e.g., nat U) used}}\\
 &= \frac{M_f}{FF\times M_l}\\
 M_f &= \mbox{mass of fuel consumed}\\
 M_l &= \mbox{mass of fuel loaded into the core}\\
 FF &= \mbox{enrichment feed factor}\\
\end{align}

## Aside: $M_f$
First, let's think about the mass of fuel question. Your book states without proof that thermal $^{235}U$ fissions consume about 1.23g per MWd. 


**Let's show that this is the case.**

First, recall :

\begin{align}
^{235}_{92}U + ^{1}_{0}n \rightarrow \left(^{236}_{92}U\right)^* \rightarrow ^{A_1}_{Z_1}X +  ^{A_2}_{Z_2}X + {\nu ^{1}_{0}n} + Q
\end{align}

That's our fission reaction. For each fission, 200MeV of energy is released. Let's then calculate how many atoms result in a thermal power of one watt.

\begin{align}
\left(\frac{1W}{200\frac{MeV}{fission}}\right)
\left(\frac{J}{W\cdot s}\right)
\left(\frac{1MeV}{1.6021\times10^{-13}J}\right) & =3.12\times 10^{10}
\left(\frac{fissions}{s}\right)
\end{align}

Each fission is one uranium atom, so:
\begin{align}
\left(\frac{3.12\times 10^{10}\mbox{atoms fissioned}}{s}\right)
\left(235 \frac{g}{g\cdot mol}\right)\left({6.022\times10^{23}\frac{atoms}{g\cdot mol}}\right)& = 1.22\times 10^{-11}\frac{g}{J}\\
\end{align}

We need to convert to MWd, so we'll need to know how many joules are in a MWd. 


\begin{align}
\left(\frac{10^6J}{MW\cdot s}\right)
\left(\frac{60\times60 s}{h}\right)
\left(\frac{24h}{d}\right)
&=
8.64\times10^{10}\frac{J}{MWd}
\end{align}

Then, we can multiply J/MWd against g/J to get g/MWd. 

\begin{align}
\left(8.64\times10^{10}\frac{J}{MWd}\right)\left(1.22\times 10^{-11}\frac{g}{J}\right)&= 1.05 \frac{g}{MWd}
\end{align}

**However, we have to correct for the fact that fuel is also consumed by non-fission capture.**

\begin{align}
\frac{\sigma_c}{\sigma_f} &= 0.169\\
\Rightarrow\frac{\sigma_f + \sigma_c}{\sigma_f} &= 1.169\\
\end{align}


So, we get:
\begin{align}
\left(\frac{g}{MWd}\right)\left(\frac{\sigma_f + \sigma_c}{\sigma_f}\right)
&= 1.05 \frac{g}{MWd}\times  1.169 = 1.23
\end{align}


```python
mev_per_fission = 200
j_per_mev = 1.60218e-13
fission_per_sec = 1/(mev_per_fission*j_per_mev)
print("fissions per second", fission_per_sec)

N_5 = 235/6.022E23
g_per_fission = fission_per_sec*N_5
print("g per fission", g_per_fission)

j_per_mwd = (10**6)*(3600)*(24)
print("Joule per MWd", j_per_mwd)

cap_to_fiss = 1 + 0.169
g_per_mwd = j_per_mwd*g_per_fission*cap_to_fiss
print("g per MWd", g_per_mwd)
```

## U Utilization

Alright, now you have 1.23 g/MWd for uranium 235 fissions. Now you can figure out the mass consumed for some burnup amount. 

\begin{align}
U &= \frac{\mbox{mass of fuel consumed}}{\mbox{mass of resource (e.g., nat U) used}}\\
 &= \frac{\frac{1.23g}{MWd}BU(tIHM)}{\mbox{mass of resource (e.g., nat U) used}}
\end{align}


Now, though, you need the feed factor, to get the denominator.

\begin{align}
FF &= \frac{x_p - x_w}{x_f - x_w}\\ 
\end{align}



\begin{align}
U &= \frac{\frac{1.23g}{MWd}BU(tIHM)}{FF\times M_l}
\end{align}

## Fuel Utilization in Breeders

A breeder reactor has the quality that it encourages the breeding of a secondariy fissile nuclide and thus, gets more fuel utilization.


Let's introduce the concept of $\gamma$, the losses in reprocessing and fuel refabrication. 

\begin{align}
U &= \frac{\frac{M_f}{M_l}}{\left[\frac{M_f}{M_l}\right](1-\gamma) + \gamma}\\
\end{align}

Let's compare once through with a breeder. 


\begin{align}
\frac{M_f}{M_l} &= \frac{g}{Mwd}\times BU\\
              &= \frac{1.23g}{MWd}\times\frac{50MWd}{tHM}\times\frac{10^3tHM}{g}\\
              &= 0.0615
\end{align}

For a typical LWR, the feed factor is :

\begin{align}
FF &= \frac{x_p - x_w}{x_f - x_w}\\ 
&= \frac{4 - .2}{0.711 - 0.2}\\
&= 7.436
\end{align}

Thus, U = 0.0615/7.436 = 0.0082 for a typical LWR.

For a reprocessing scheme with 2% losses:


\begin{align}
U &= \frac{0.0615}{\left[0.0615\right](1-0.02) + 0.02}\\
\end{align}

```python
U_lwr = 50*1.23/(1000*7.436)
U_breed = (50*1.23/(1000)) / ((0.0615)*(1-0.02) + 0.02)
print("u_lwr", U_lwr)
print("u_breed", U_breed)
```

### Aside: Delayed Neutron Fraction and Mean Neutron Lifetime

![./xs.jpg](./xs.jpg)

By combining the neutron transport equation with a source contribution from
delayed neutrons,  we can arrive at the fundamental analysis tool in nuclear reactor
kinetics. These are a stiff set of PDEs involving nonlinearly coupled
neutronics, thermodynamics, and hydraulics.

The delayed neutron precursors obey the equation,

\begin{align}
  \frac{\partial \hat{C}_i(t,r)}{\partial t} =
  \beta_i\nu\Sigma_f(r,t)\phi(r,t) -
  \lambda_i\hat{C}_i(r,t)\\
  i\in [1,6].
\end{align}

$\beta_i$ and $\lambda_i$ depend on the incident neutron energy, fissionable
isotope, and precursor group.  In this way, the  linearized Boltzmann transport
equation has seven dimensions.  Taking delayed neutrons into account, the one
speed neutron diffusion equation can be written,

\begin{align}
  \frac{1}{v}\frac{\partial \phi(r,t)}{\partial t} - D(r,t)\nabla^2\phi(r,t) + \
  \Sigma_a(r,t)\phi(r,t) = (1-\beta)\nu\Sigma_F(r,t)\phi(r,t) + \
  \sum_{i=1}^6\lambda_i\hat{C}_i(r,t).
\end{align}

Transient analysis methods seek to solve this equation for changes in the
parameters caused by a changing reactor conditions. Additional PDEs are added to
this calculation to capture the dependence of temperature on heat conduction and
the effect of fluid flow.


\begin{equation}
  \frac{d}{dt}\left[
    \begin{array}{c}
      p\\
      \zeta_1\\
      .\\
      .\\
      .\\
      \zeta_J\\
      \omega_1\\
      .\\
      .\\
      .\\
      \omega_K\\
    \end{array}
    \right]
    =
    \left[
      \begin{array}{ c }
        \frac{\rho(t,T_{fuel},T_{cool},\cdots)-\beta}{\Lambda}p +
        \displaystyle\sum^{j=J}_{j=1}\lambda_j\zeta_j\\
        \frac{\beta_1}{\Lambda} p - \lambda_1\zeta_1\\
        .\\
        .\\
        .\\
        \frac{\beta_J}{\Lambda}p-\lambda_J\zeta_J\\
        \kappa_1p - \lambda_1\omega_1\\
        .\\
        .\\
        .\\
        \kappa_{k}p - \lambda_k\omega_{k}\\
      \end{array}
      \right]
    \end{equation}




\begin{align}
    \rho(t,&T_{fuel},T_{cool},T_{mod}, T_{refl}) = \mbox{ reactivity, [pcm]}\\
    \beta &= \mbox{ fraction of neutrons that are delayed}\\
    \beta_j &= \mbox{ fraction of delayed neutrons from precursor group j}\\
    \zeta_j &= \mbox{ concentration of precursors of group j}\\
    \lambda^{d}_j &= \mbox{ decay constant of precursor group j}\\
    \omega_k &= \mbox{ decay heat from FP group k}\\
    \kappa_k &= \mbox{ heat per fission for decay FP group k}\\
    \lambda^{FP}_k &= \mbox{ decay constant for decay FP group k}\\
    l &= \mbox{mean neutron lifetime}\\
      &= \mbox{mean time for one neutron to be removed from the reactor}\\
      &= \frac{1}{\left(\Sigma_a + DB^2\right)v}\\
    \Lambda &= \mbox{neutron reproduction time}\\
            &= \mbox{mean time for one neutron to be replaced by another neutron on fissioning}\\
            &= \frac{1}{\nu\Sigma_fv}\\
 \end{align}

**Bottom line: lower fraction of delayed neutrons constrains reactor control.**
**Bottom line: Similarly, shorter prompt neutron lifetimes neutrons constrain reactor control.**

# Conversion Ratio

The conversion ratio is a measure of the evolution of fissile nuclides in the core and their involvement in fission.

\begin{align}
CR &= \frac{\mbox{production of fissile nuclei}}{\mbox{depletion of fissile nuclei}}\\
&=\begin{cases}
< 1 & \mbox{burner} \\
> 1 & \mbox{breeder} 
\end{cases}
\end{align}

## Breeding in Thermal Reactors

For typical enriched uranium fuel, this is, effectively:

\begin{align}
CR &= \frac{\mbox{production of } ^{239}Pu}{\mbox{depletion of } ^{235}U}
\end{align}


If $\eta_0$ is the number of neutrons produced for each neutron absorbed in $^{235}U$ is 2.075, then what has to happen in order to have a $^{235}U$ fuelled breeder?

- 1st neutron: creates another fission 
- 2nd neutron: produces ${239}Pu$
- 0.08 neutrons left over

Is the above situation likely?


There is an opportunity for thermal spectrum using $^{233}U$, which has a higher $\eta_0$, equal to 2.296. Of course, the way this works is that $^{232}Th$ is the fertile nuclide and $^{233}U$ is the fissile fuel that is preferentially bred. However, one must start with $^{235}U$. 


## Breeding in Fast Reactors

We can do a bit better with a plutonium-based fuel cycle. Since ${239} Pu$ has a high capture to fission cross section at thermal energies, we can only do this by aiming for a harder (faster, less thermal, higher energy) neutron spectrum in the core.

- In a fast reactor of this sort, the neutrons are not thermalized. Instead, $^{239}Pu$ fissions occur at high energies (above 1MeV). 
- At these energies, the fission cross section in $^{239}Pu$ is $\sigma_f = 1.7b$ 
- and the capture-to-fission ratio is less than 0.03. 
- Also, at about 3 neutrons per fission for ${239} Pu$ fissions at 1MeV, there are plenty of neutrons available for breeding.

### LMFBR Coolant
Fast reactors must avoid thermalization of the neutrons. Thus, moderating coolants are avoided. Instead, liqiud metals are chosen as the coolant for their relatively high mass number A. An added advantage of liquid metals is that they are excellent at heat transfer.
